```python
def adicao ():
  num1 = (float(input('Digite o primeiro valor da soma: ')))
  num2 = (float(input('Digite o segundo valor da soma: ')))

  soma = num1 + num2
  print (f'O valor da soma dos dois valores é de {soma}!')

def subtracao():
  num1 = (float(input('Digite o primeiro valor da subtração: ')))
  num2 = (float(input('Digite o segundo valor da subtração')))

  sub = num1 - num2
  print (f'O valor da subtração desses dois números é de {sub}!')

def divisao ():
  num1 = (float(input('Digite o primeiro valor da divisão: ')))
  num2 = (float(input('Digite o segundo valor da divisão: ')))

  div = num1/num2
  print (f'O valor da divisão dos dois números é de: {div}!')

def multiplicacao ():
  num1 = (float(input('Digite o primeiro número da Multiplicação: ')))
  num2 = (float(input('Digite o segundo número da Multiplicação')))

  mult = num1*num2
  print (f'O resultado da sua multiplicação é: {mult}!')

def calculadora ():

  while True:
        print ('Bem vindo á sua calculadora!!')
        print ('Escolha entre as opções abaixo para seguir com os cálculos: ')
        print ('1 - Adição de dois números')
        print ('2 - subtração de dois números')
        print ('3 - divisão de dois números')
        print ('4 - multiplicação de dois números')

        escolha = (input("Escolha a opção desejada ou aperte 'Q' par sair: "))
        if escolha == 'q':
            print ('Saindo da calculadora')
            break

        elif escolha == '1':
            adicao()
            print ('Deseja realizar um novo cálculo?')
            resposta = (input("Digite sim ou não: "))
            if resposta == 'sim':
                calculadora()
            else:
                resposta == 'não'
                break

        elif escolha =='2':
            subtracao()
            print ('Deseja realizar um novo cálculo?')
            resposta = (input("Digite sim ou não: "))
            if resposta == 'sim':
                calculadora()
            else:
                resposta == 'não'
                break
        elif escolha == '3':
            divisao()
            print ('Deseja realizar um novo cálculo?')
            resposta = (input("Digite sim ou não: "))
            if resposta == 'sim':
                calculadora()
            else:
                resposta == 'não'
                break
        elif escolha == '4':
            multiplicacao()
            print ('Deseja realizar um novo cálculo?')
            resposta = (input("Digite sim ou não: "))
            if resposta == 'sim':
                calculadora()
            else:
                resposta == 'não'
                break
        else:
            print ('Opção inválida!')
calculadora()
```
