```python
itens = []
quantidades = []

def adicionar_item (nome, quantidade):
    if nome in itens:
        index = itens.index(nome) #comando .index serve para retornar o indice de um item da lista
        quantidades[index] += quantidade
    else:
        itens.append(nome) #comando .append adiciona um elemento no final da lista
        quantidades.append(quantidade)
    print ('---------------------------------------------')    
    print ('Item adicionado com sucesso!')
    print ('---------------------------------------------')  
    
def remover_itens (nome):
    if nome in itens:
        index= itens.index(nome)
        del itens[index]
        del quantidades[index]
        print ('---------------------------------------------')
        print ('item removido com sucesso!')
        print ('---------------------------------------------') 
    else:
        print ('Não temos esse item em estoque')
        
def verificar_quantidade(nome):
    if nome in itens:
        index = itens.index(nome)
        print ('---------------------------------------------')
        print (f'quantidade de {nome} no estoque é de: {quantidade[index]} unidades')
        print ('---------------------------------------------') 
    else:
        print ('---------------------------------------------')
        print ('o item não existe no estoque')
        print ('---------------------------------------------')
        
def exibir_estoque():
        print ('---------------------------------------------')
        print ('Estoque atual: ')
        for i in range(len(itens)):
            print (f'{itens[i]}: {quantidade[i]} unidades')
            print ('---------------------------------------------')         
        
while True:
    print ('Olá, seja bem vindo ao seu sistema de estoque')
    print ('1 - Adicionar item ao estoque')
    print ('2 - Remover item do estoque')
    print ('3 - Verificar quantidade de itens no estoque: ')
    print ('4 - Exibir estoque')
    print ("3 - press 'q' para sair")
    resposta = (input('Digite o número da ação desejada: '))
    if resposta == '1':
        nome = (input('Digite o nome do produto a ser adicionado: '))
        quantidade = (input('Digite a quantide do produto a ser adicionado: '))
        adicionar_item(nome,quantidade)
    elif resposta == '2':
        nome = (input('Digite o nome do item a ser removido: '))
        remover_itens(nome)
    elif resposta == '3':
        nome = (input('Digite o nome do item que deseja verificar a quanntidade em estoque: '))
        verificar_quantidade(nome)
    elif resposta == '4':
        exibir_estoque()
        exibir_estoque()    
    elif resposta == 'q':
        print('Obrigado por utilizar o estoque!')
        break
    else:
        print ('Opção inválida, tente novamente!')
            
    
            
```
